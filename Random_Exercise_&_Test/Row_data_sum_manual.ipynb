```python
import numpy as np
import pandas as pd
```

```python
pandas_df = pd.DataFrame(data = np.random.randint(0, 500, (15,20)))
```

```python
pandas_df
```

```python
#Universal DataFrame Sumup by columns by Dr.AAA

pandas_df["H-Sum"] = 0


for rows in np.arange(0, (pandas_df.shape[0])):
    cols_sum = 0
    for cols in pandas_df.iloc[rows,]:
    
        cols_sum = cols_sum + cols
   
    #print(cols_sum)    
    pandas_df.iloc[rows ,(pandas_df.shape[1] - 1)] = cols_sum
```
