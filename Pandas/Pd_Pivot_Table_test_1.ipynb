```python
df.groupby("sex")[["survived"]].mean()
```

```python
df.groupby(["sex", "embark_town"])[["survived"]].mean()
```

```python
df.groupby(["sex", "embark_town", "class"])[["survived", "age"]].mean()
```

```python
df.groupby(["sex", "embark_town", "class"])[["survived"]].aggregate(["mean", "sum", "std"])
```

```python
df.groupby(["sex", "embark_town", "class"])[["survived", "age"]].mean().unstack()
```

```python
df.pivot_table(values = ["survived", "age"], index = ["sex", "embark_town"], columns = ["class"])
```

```python
df.pivot_table(values = ["survived", "age"], index = ["sex", "embark_town"], columns = ["class"], aggfunc = ["sum", "mean", "std"])
```

```python
df.pivot_table(values = ["survived"], index = ["sex"], columns = ["class"], aggfunc = ["sum", "mean", "std"])
```
