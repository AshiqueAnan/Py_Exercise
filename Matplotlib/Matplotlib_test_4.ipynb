```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
X = np.arange(1,16)
Y = np.arange(100,1600, 100)
```

```python
Figure, Axes = plt.subplots(nrows = 2, ncols = 2)
```

```python
print(Axes[0])
```

```python
Figure, Axes = plt.subplots(nrows = 2, ncols = 1)

Axes[0].plot(X,Y)
Axes[0].set_title("G1")

Axes[1].plot(X,X**2)
Axes[1].set_title("G2")

plt.tight_layout()
```

```python
Axes.size
print(Axes)
```

```python
Figure_1, Axes_1 = plt.subplots(nrows = 2, ncols = 2)

Axes_1[0][0].plot(X,Y)
Axes_1[0][0].set_title("G1")

Axes_1[0][1].plot(X,X**2)
Axes_1[0][1].set_title("G2")

Axes_1[1][0].plot(Y, np.exp(X))
Axes_1[1][0].set_title("G3")

Axes_1[1][1].plot(X, np.sqrt(X))
Axes_1[1][1].set_title("G4")

plt.tight_layout()
```
