```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
X = np.arange(0, math.pi*4, 0.05)
```

```python
Y = np.sin(X)
```

```python
Fig = plt.figure()
```

```python
Axis = Fig.add_axes([0,0,1,1])
```

```python
Axis.plot(X,Y, color = "red")
Axis.set_xticks([0,2,4,6,8,10,12,14])
Axis.set_yticks([-1,0,1])
Axis.spines["top"].set_visible(False)
Axis.spines["right"].set_visible(False)
Axis.grid(True)
```
