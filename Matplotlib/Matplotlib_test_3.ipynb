```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
X = np.arange(-1,5,0.1)
```

```python
Y = np.exp(-X)*np.cos(5*X)
```

```python
plt.plot(X,Y)
plt.show()
```

```python
plt.figure(figsize = (10,10))
```

```python
plt.figure(figsize = (10,8))
plt.plot(X,Y)
plt.show()
```

```python
plt.figure(figsize = (10,8), dpi = (120))
plt.stackplot(X,Y)
plt.show()
```

```python
x1 = np.arange(1,16)
```

```python
y1 = np.arange(10,160,10)
y1
```

```python
plt.figure(figsize = (5,4), dpi = 150)
plt.plot(x1, y1)
plt.show()
```

```python
plt.subplot(2, 3, 1)
plt.plot(x1, x1*x1, "red")

plt.subplot(2, 3, 2)
plt.plot(x1, x1*x1*x1, "green")

plt.subplot(2, 3, 3)
plt.plot(x1, np.exp(x1), "yellow")

plt.subplot(2, 3, 4)
plt.plot(x1, np.log10(x1), "blue")

plt.subplot(2, 3, 5)
plt.plot(x1, np.sqrt(x1), 'bo')

plt.subplot(2, 3, 6)
plt.plot(x1, (x1/np.sqrt(x1)), 'go')

plt.show()
```
