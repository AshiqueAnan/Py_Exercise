```python
import numpy as np
import matplotlib.pyplot as plt

#for ploting graphs in Jupyterlab
%matplotlib inline
```

```python
X_axis = np.arange(-500, 500)
```

```python
Y_axis = np.random.randint(-1000, 1000, 1000)
```

```python
plt.plot(X_axis, Y_axis)
```

```python
Y_sort = np.sort(Y_axis)
```

```python
plt.plot(X_axis, Y_sort)
plt.show()
```

```python
Y_sqr = X_axis**2
```

```python
plt.plot(X_axis, Y_sqr)
plt.show()
```

```python
plt.plot(X_axis, Y_sqr,linestyle='dashed')

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("GFX")
plt.show()
```

```python
x_axis = np.linspace(-10,10, num = 5000)

plt.plot(x_axis, np.sin(x_axis), "r-")
plt.plot(x_axis, np.cos(x_axis), "g--")
plt.show()
```
