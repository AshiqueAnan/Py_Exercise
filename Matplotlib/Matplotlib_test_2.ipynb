```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x_ax = np.arange(1,100)
```

```python
y_ax = x_ax**2
```

```python
plt.plot(x_ax, y_ax)
plt.show()
```

```python
plt.bar(x_ax, y_ax)
plt.show()
```

```python
plt.barh(x_ax, y_ax)
plt.show()
```

```python
plt.polar(x_ax, y_ax)
plt.show()
```

```python
plt.step(x_ax, y_ax)
plt.show()
```
