```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
Data_flight = pd.read_csv("flight_details.csv")
```

```python
Data_flight
```

```python
Sorted_data = Data_flight.pivot(index = "Months", columns = "Years", values = "Passengers")
```

```python
Sorted_data
```

```python
sns.set()
plt.figure(dpi = 120)
sns.heatmap(data = Sorted_data)
plt.title("Heatmap Data of Flights")
plt.show()
```

```python
sns.set()
plt.figure(dpi = 120)
sns.heatmap(data = Sorted_data, cmap = sns.light_palette("blue", as_cmap = True , reverse = True))
plt.title("Heatmap Data of Flights In BLUE MFFFF")
plt.show()
```
