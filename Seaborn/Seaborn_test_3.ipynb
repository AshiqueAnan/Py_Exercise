```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
Data_IQ = pd.read_csv("scores.csv")
```

```python
Data_IQ
```

```python
fig = plt.figure(dpi = 120)
axis = fig.add_axes([0,0,1,1])

sns.boxplot(data = Data_IQ)
axis.set(xlabel = "Group", ylabel = "IQ Scores")
sns.set_style("whitegrid")
plt.title("IQ score of some group of people")
plt.show()
```

```python
fig.savefig("IQ-Scores.png", bbox_inches="tight")
```
